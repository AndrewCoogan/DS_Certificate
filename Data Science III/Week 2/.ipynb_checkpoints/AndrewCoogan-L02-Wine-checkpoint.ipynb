```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from scipy.io import arff
from sklearn.naive_bayes import GaussianNB
from itertools import combinations
```

```python
wine = pd.read_csv("RedWhiteWine.csv")
wine_a = pd.DataFrame(arff.loadarff("RedWhiteWine.arff")[0])
```

This data is the same, I am going to use the csv as its formatted more friendly than the arff file that has been loaded as a binary.  I have tried to convert it, but I am having issues converting to strings.

```python
target = ['Class']
classifiers = ['free sulfur dioxide', 'total sulfur dioxide', 'citric acid', 'residual sugar', 'chlorides', 'density']

feature_train, feature_test, target_train, target_test = train_test_split(
    wine[classifiers], 
    wine[target], 
    random_state=0,
    test_size = 0.2) 
```

```python
print('Total number: ', len(wine.quality), '\nTraining set : ', len(target_train), '\nTesting set: ', len(target_test))
```

#### Train And Test A Naive Bayesian Classification Model Using All Variables

```python
gnb = GaussianNB()
gnb_model = gnb.fit(feature_train, target_train)
y_pred = gnb_model.predict(feature_test)
```

```python
print('Accuracy: ', np.round(100 * (y_pred == target_test.Class).sum()/len(target_test),2), '%')
```

Lets run and check all possible subsets...

```python
best_accuracy = 0.0
best_combination = list()

for i in range(1, len(classifiers)+1):
    comb = combinations(classifiers, i)
    for c in comb:
        c_list = list(c)
        gnb_l = GaussianNB()
        gnb_model_l = gnb_l.fit(feature_train[c_list], target_train)
        y_pred_loop = gnb_model_l.predict(feature_test[c_list])
        accuracy = np.round(100 * (y_pred_loop == target_test.Class).sum()/len(target_test),2)
        if accuracy > best_accuracy: 
            best_accuracy = accuracy
            best_combination = c_list
        print('Variables used: ', ', '.join(c_list))
        print('Accuracy: ', accuracy, '%')

print('------------------------------------------')
print('Best combination: ', ', '.join(best_combination))
print('Best accuracy: ', best_accuracy, '%')
```

Looping over all iterations (we are using a small subset of the larger data, so this is not too computationally taxxing), a combination of total sulfur dioxide, citric acid, resudual sugar, chlorides, and density prove to train the best model, even better model than using all six classifiers.  The optimal model removed free sulfur dioxide.  We are able to get an accuracy of 97.23% versus the training data, using a 80:20 training testing ratio.  In the above loop I iterate all possible combinations including single classifiers and including all classifiers (even though I did it above, I reran it so that we can see the whole picture).

The objective of this assignment was to achieve a 95%+ using the minimum number of classifiers.  We are able to achieve this by using two variables: total sulfur dioxide and density, getting 95.31% accuracy.
