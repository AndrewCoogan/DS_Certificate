Classifying wines using decision tree classifiers given a set of parameters

```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from itertools import combinations
from IPython.display import Image
from subprocess import check_call
```

```python
wine = pd.read_csv("RedWhiteWine-1.csv")
wine.head()
```

The data looks pretty straight forward.  All the columns we will use (defined below) are continuous variables and the class is either 1 or 2 (red or white).  We will use these variables without normalization to see if we can determine what the class each wine is.  We will separate the dataset into a testing and training with the proportions being 20% and 80% respectively.

```python
target = ['Class']
classifiers = ['free sulfur dioxide', 'total sulfur dioxide', 'sulphates', 'pH', 'alcohol', 'density']

feature_train, feature_test, target_train, target_test = train_test_split(
    wine[classifiers], 
    wine[target], 
    random_state = 0,
    test_size = 0.2)
```

### Entropy Test First ###

```python
dtree_entropy = DecisionTreeClassifier(criterion = 'entropy', max_depth = 3)
model_dt_entropy = dtree_entropy.fit(feature_train, target_train)

# Validate the model
y_predict_entropy = model_dt_entropy.predict(feature_test)
accuracy_entropy = accuracy_score(target_test, y_predict_entropy) * 100
print("Accuracy is : {}%".format(accuracy_entropy))
```

```python
dotfile1 = open("all_var_model.dot", 'w')
tree.export_graphviz(model_dt_entropy, out_file = dotfile1, filled=True, rounded=True, feature_names = classifiers)
dotfile1.close()

# Convert the dot file to a png
check_call(['dot','-Tpng','all_var_model.dot','-o','all_var_model.png'])

Image(filename='all_var_model.png', width=500, height=500)
```

Cool--Looks like using all variables and a max depth of 3, we can get a 96.38% accuracy.

```python
best_accuracy = 0.0
best_combination = list()

for i in range(1, len(classifiers)+1):
    comb = combinations(classifiers, i)
    for c in comb:
        c_list = list(c)
        dtc_l = DecisionTreeClassifier(criterion = 'entropy', max_depth = 6)
        #  We have to increase the max_depth or else it will just use the three variables I describe below.
        #  I chose six as that would allow for all variables to be used
        dtc_model_l = dtc_l.fit(feature_train[c_list], target_train)
        y_pred_loop = dtc_model_l.predict(feature_test[c_list])
        accuracy = accuracy_score(target_test, y_pred_loop) * 100
        if accuracy > best_accuracy: 
            best_accuracy = accuracy
            best_combination = c_list
        print('Variables used: ', ', '.join(c_list))
        print('Accuracy: ', accuracy, '%')

print('------------------------------------------')
print('Best combination: ', ', '.join(best_combination))
print('Best accuracy: ', best_accuracy, '%')
```

Below I will go through three examples that were run in the loop above.

```python
#  This example is the solution that resulted in >95% accuracy using the fewest variables
#  According to our test, this is the best way to determine if a wine is a 
#  red or a white given the classifiers available.

sub_features = ['total sulfur dioxide', 'density']
dtc = DecisionTreeClassifier(criterion = 'entropy', max_depth = 3)
dtc_model = dtc.fit(feature_train[sub_features], target_train)
y_pred_loop = dtc_model.predict(feature_test[sub_features])
accuracy = accuracy_score(target_test, y_pred_loop) * 100
print('Accuracy ', np.round(accuracy, 3), '%')

dotfile1 = open("two_var_model.dot", 'w')
tree.export_graphviz(dtc_model, out_file = dotfile1, filled=True, 
                     rounded=True, feature_names = sub_features)
dotfile1.close()

# Convert the dot file to a png
check_call(['dot','-Tpng','two_var_model.dot','-o','two_var_model.png'])

Image(filename='two_var_model.png', width=500, height=500)
```

```python
#  This example had the LOWEST accuracy for utilizing three varibales.

sub_features = ['sulphates', 'pH', 'alcohol']
dtc = DecisionTreeClassifier(criterion = 'entropy', max_depth = 3)
#  I lowered the max_depth to three just so it would be easier to see the image.
dtc_model = dtc.fit(feature_train[sub_features], target_train)
y_pred_loop = dtc_model.predict(feature_test[sub_features])
accuracy = accuracy_score(target_test, y_pred_loop) * 100
print('Accuracy ', np.round(accuracy, 3), '%')

dotfile1 = open("three_var_model.dot", 'w')
tree.export_graphviz(dtc_model, out_file = dotfile1, filled=True, 
                     rounded=True, feature_names = sub_features)
dotfile1.close()

# Convert the dot file to a png
check_call(['dot','-Tpng','three_var_model.dot','-o','three_var_model.png'])

Image(filename='three_var_model.png', width=500, height=500)
```

```python
#  This example had the GREATEST accuracy for utilizing any nuumber of variables
#  The decision tree decideded to not inlcude pH at all.
#  The forest in the image below is kind of a mess with more information than I can really follow.

sub_features = ['free sulfur dioxide', 'total sulfur dioxide', 'sulphates', 'alcohol', 'density']
dtc = DecisionTreeClassifier(criterion = 'entropy', max_depth = 6)
#  I left this one at six so that I could replicate the loop
dtc_model = dtc.fit(feature_train[sub_features], target_train)
y_pred_loop = dtc_model.predict(feature_test[sub_features])
accuracy = accuracy_score(target_test, y_pred_loop) * 100
print('Accuracy ', np.round(accuracy, 3), '%')

dotfile1 = open("five_var_model.dot", 'w')
tree.export_graphviz(dtc_model, out_file = dotfile1, filled=True, 
                     rounded=True, feature_names = sub_features)
dotfile1.close()

# Convert the dot file to a png
check_call(['dot','-Tpng','five_var_model.dot','-o','five_var_model.png'])

Image(filename='five_var_model.png', width=500, height=500)
```
