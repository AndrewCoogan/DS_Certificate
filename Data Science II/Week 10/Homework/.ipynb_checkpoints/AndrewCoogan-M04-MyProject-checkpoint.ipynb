# Abstract
Magic: The Gathering is a card game that is a combination of resource management and creativity where there are numerous win conditions.  With every new set that is released, new cards are added to the universe and new win conditions are discovered.  In competitive settings there are even banned lists which add special dynamics to the game play.  What I want to focus on are certain keywords used in the cards (deathtouch, lifelink, etc.) and combinations of those and see if the cost of those cards can be projected using other variables.

## Problem
As beiefly stated above, there are keywords used on certain kinds of cards (I will elaborate on these as we move on and give very basic introduction for some understanding) and I want to see if there are correlations and dependencies on these keywords.

## Conclusion
The magnitude of power * toughness is the best metric for determining converted mana cost.  There are some correlations otherwise but a lot of the factors did not contribute constructively to the models.  The nature of magic the gathering is that each of the colours and types are balanced, so they should share a similar mana curve.  Perhaps for future research, more subsections of the data should be looked into and see if correlation within types and other classifiers can be more telling.  There are also proven neural networks for the creation of new cards.  That approach could be utilized for extracting more hidden layers.

```python
# Load necessary libraries
import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as sm
import seaborn as sns
from sklearn.decomposition import PCA
%matplotlib inline
```

```python
cards = pd.read_json(data_location, orient = "index")
cards.head()
```

There have been sets released such as unhinged or unglued which have 'silly' cards.  These cards do not see normal play as they have odd properties such as infinite mana (read, it costs infinity to cost it), or fractional power and toughness (read, how much damage it can deal/how much damage it can take before being destroyed).  Here I will filter out these "outliers".

```python
cards_f = cards
cards_f = cards_f.drop(cards[cards.cmc > 30].index)
cards_f = cards_f.drop(cards[cards.layout.str.contains('token', re.IGNORECASE)].index)
cards_f = cards_f.drop(cards[cards.layout.str.contains('scheme', re.IGNORECASE)].index)
cards_f = cards_f.drop(cards[cards.layout.str.contains('plane', re.IGNORECASE)].index)
cards_f = cards_f.drop(cards[cards.layout.str.contains('vanguard', re.IGNORECASE)].index)
cards_f = cards_f.drop(cards[cards.layout.str.contains('phenomenon', re.IGNORECASE)].index)
#cards_f = cards_f.drop(cards.loc[cards.types.str.join('|').str.contains('conspiracy', re.IGNORECASE, na = False).index].index)
#cards_f = cards_f.drop(cards.loc[cards.types.str.join('|').str.contains('host', re.IGNORECASE, na = False).index].index)
cards_f = cards_f.loc[~(cards.cmc == 0.5)]
cards_f = cards_f.loc[~(cards.power == '*Â²')]
cards_f = cards_f.loc[~(cards.power == 'Â½')]
cards_f = cards_f.loc[~(cards.power == '1Â½')]
cards_f = cards_f.loc[~(cards.toughness == '1Â½')]
cards_f = cards_f.loc[~(cards.power == '2Â½')]
cards_f = cards_f.loc[~(cards.power == '3Â½')]
cards_f = cards_f.loc[~(cards.toughness == '3Â½')]
cards_f = cards_f.loc[~(cards.power == '99')]
cards_f = cards_f.loc[~(cards.power == '+1')]
cards_f = cards_f.loc[~(cards.power == '+2')]
cards_f = cards_f.loc[~(cards.power == '+3')]
cards_f = cards_f.loc[~(cards.power == '+4')]
cards_f = cards_f.loc[~(cards.power == 'âˆž')]
cards_f = cards_f.loc[~(cards.power == '?')]
```

There are some extra columns we dont really care about, so I am only going to keep a subset of the columns.

```python
cols_to_keep = ['cmc', 'colorIdentity', 'loyalty', 'manaCost', 'power','toughness','text',
                'subtypes', 'supertypes', 'types']
cards_f = cards_f[cols_to_keep]
```

```python
colors = cards_f.colorIdentity.str.join('|').str.get_dummies()
types = cards_f.types.str.join('|').str.get_dummies()
#sub_types = cards_f.subtypes.str.join('|').str.get_dummies()
#super_types = cards_f.supertypes.str.join('|').str.get_dummies()
```

```python
picked_text_keywords = ['deathtouch', 'lifelink', 'dredge', 'storm', 'annihilator', 'hexproof', 'shroud', 'flying',
                       'proliferate', 'vigilance', 'defender', 'haste', 'reach', 'flash', 'indestructible', 'delve',
                       'affinity']
picked_text_keywords = [str.upper(t) for t in picked_text_keywords]

keywords = pd.DataFrame(index = cards_f.index)

for k in picked_text_keywords:
    temp_series = pd.Series(index=cards_f.index)
    temp_series[:] = 0
    temp_series.loc[cards_f.loc[cards_f.text.str.contains(k, case = False, na = False)].index] = 1
    keywords[k] = temp_series
```

```python
frames = [colors, types, cards_f.cmc, cards_f.loyalty, cards_f.power, cards_f.toughness, keywords]
frames = [pd.DataFrame(item) for item in frames]
cards_d = pd.concat(frames, axis = 1)
```

##  Data Cleaning Done!

Lets take a quick look and see if we can get a basic linear regression to predict cmc using all classifiers.  But first lets see what the density of our target, cmc (converted mana cost), looks like.

```python
plt.yscale('log', nonposy='clip')
plt.xlabel('Converted Mana Cost (cmc)')
plt.ylabel('Log Count')
plt.title('Histogram of CMC')
plt.hist(cards_d.cmc, bins = len(set(cards_d.cmc)))
```

There are a lot of cards to take into account, and only a very few at the top end of the spectrum (12+ cmc), so I did a log on the y-axis.  Between 2 and 12 there is actually a very nice downward logarithmic curve.

```python
classifiers = cards_d.columns.tolist()
target_label = 'cmc'
classifiers.remove(target_label)

formula_terms = ' + '.join(classifiers)
formula = target_label + ' ~ ' + formula_terms

ols_model = sm.ols(formula = formula, data = cards_d)

results = ols_model.fit()
n_points = cards_d.shape[0]
y_output = cards_d[target_label].reshape(n_points, 1)

# Get slope (m) and y-intercept (b)
print('Intercept, Slopes : \n{}'.format(results.params))

# Get the t-values (hypothesis test statistics) for linear regression coefficient hypothesis tests.
print('Intercept t-value, Slope t-values: \n{}'.format(results.tvalues))

# Get p-values for above t-value statistics
print('\nHypothesis test summary for each coefficient if they differ from zero:')
print(results.pvalues)

print('\nSSE, SST, SSR, and RMSE:')
mean_y = np.mean(y_output)
sst = np.sum((y_output - mean_y)**2)
sse = sst - results.ssr
print('SSE: {}'.format(sse))
print('SST: {}'.format(sst))
print('SSR: {}'.format(results.ssr))
print('MSE: {}'.format(results.mse_model))
print('RMSE: {}'.format(np.sqrt(results.mse_model)))

# Get most of the linear regression statistics we are interested in:
print(results.summary())

# Plot a histogram of the residuals
sns.distplot(results.resid, hist=True)
plt.xlabel('Residual')
plt.ylabel('Frequency')
plt.title('Residual Histogram')
```

Not a bad place to start with an Adj R-Squared of 0.622!  This uses all vaiables that have been slighly trimmed.  Lets see if using just the types (Artifact, Creature, Enchantment, etc), can make anything happen.  I am going to hypothezise no, becuase there is a wide cmc band for each of the types.

I am going to look at PCA quickly.  For this I am going to scale the cmc, and only look at cards with numeric power and toughness (ie, removing the 1+ / * / etc).  Once we get rid of that, we can scale the power and toughnesses.  

```python
cards_s = cards_d.copy()
cards_s['cmc'] = (cards_s['cmc'] - np.nanmean(cards_s.cmc, )) / np.nanstd(cards_s.cmc)
cards_s = cards_s[pd.to_numeric(cards_s.power, errors='coerce').notnull()]
cards_s = cards_s[pd.to_numeric(cards_s.toughness, errors='coerce').notnull()]
cards_s['power'] = pd.to_numeric(cards_s.power, errors='coerce')
cards_s['toughness'] = pd.to_numeric(cards_s.toughness, errors='coerce')
cards_s['power'] = (cards_s['power'] - np.nanmean(cards_s.power, )) / np.nanstd(cards_s.power)
cards_s['toughness'] = (cards_s['toughness'] - np.nanmean(cards_s.toughness, )) / np.nanstd(cards_s.toughness)
cards_s = cards_s.fillna(0)
```

```python
classifiers = cards_s.columns.tolist()
target_label = 'cmc'
classifiers.remove(target_label)

pca = PCA(n_components = len(cards_s.columns.tolist()) - 1)
pca_result = pca.fit_transform(cards_s[classifiers])
pca_df = pd.DataFrame(data = pca_result, columns = classifiers)
pca_df[target_label] = cards_s.cmc
```

PCA Complete!  Lets see how the residuals look.

```python
plt.plot(pca_df.iloc[:,2:len(classifiers)])
plt.title('Explained variance by Principal Component Num')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance')
```

The confidence intervals are positive for all three of these which means that they tend to be on more expensive creatures.
